## Untils function

```python
def get_rotation_matrix(pitch, yaw, roll):

    roll = np.deg2rad(roll)
    pitch = np.deg2rad(pitch)
    yaw = np.deg2rad(yaw)

    Rx = np.array([
        [1,0,0],
        [0,np.cos(roll),-np.sin(roll)],
        [0,np.sin(roll), np.cos(roll)]
    ])

    Ry = np.array([
        [np.cos(pitch),0,np.sin(pitch)],
        [0,1,0],
        [-np.sin(pitch),0,np.cos(pitch)]
    ])

    Rz = np.array([
        [np.cos(yaw),-np.sin(yaw),0],
        [np.sin(yaw), np.cos(yaw),0],
        [0,0,1]
    ])

    return Rz @ Ry @ Rx

def compute_bbox_corners(location, extent, rotation):

    x,y,z = location
    ex,ey,ez = extent
    roll,pitch,yaw = rotation

    corners = np.array([
        [ ex, ey, ez],
        [-ex, ey, ez],
        [-ex,-ey, ez],
        [ ex,-ey, ez],
        [ ex, ey,-ez],
        [-ex, ey,-ez],
        [-ex,-ey,-ez],
        [ ex,-ey,-ez]
    ])

    R = get_rotation_matrix(roll,pitch,yaw)

    corners = corners @ R.T
    corners += np.array([x,y,z])

    return corners
```

## Compute corners

```python
def generate_car_corners_list(index):
  # load yaml config file
  yaml_path = os.path.join(DATASET_PATH, f"{index}.yaml")
  with open(yaml_path, "r") as f:
      yaml_data = yaml.safe_load(f)

  # compute bbox corners
  car_corners_list = []
  for vid, vehicle in yaml_data["vehicles"].items():
      location = np.array(vehicle["location"]) + np.array(vehicle["center"])
      extent = vehicle["extent"]
      rotation = vehicle["angle"]

      corners = compute_bbox_corners(location, extent, rotation)
      car_corners_list.append(corners)

  # save bbox corners to car box file
  # (N_vehicles * 8 * 3) npy file
  car_corners_list = np.array(car_corners_list)
  labels_radar_save_path = os.path.join(MYDATASET_CAR_BOX_PATH, f"{index}.npy")
  np.save(labels_radar_save_path, car_corners_list)
```

```python
index_range = ["016653", "017853"]
index = index_range[0]

count = 0
while index != index_range[1]:
    generate_car_corners_list(index)

    index = f"{int(index) + 1:06d}"
    count += 1
    if count % 10 == 0:
        print(index)
```
