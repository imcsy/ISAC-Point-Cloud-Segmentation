```python
import numpy as np
import open3d as o3d
import os
```

## Untils function

```python
def get_rotation_matrix(lidar_rot):
    roll, pitch, yaw = lidar_rot

    roll = np.deg2rad(roll)
    pitch = np.deg2rad(pitch)
    yaw = np.deg2rad(yaw)

    Rx = np.array([
        [1,0,0],
        [0,np.cos(roll),-np.sin(roll)],
        [0,np.sin(roll), np.cos(roll)]
    ])

    Ry = np.array([
        [np.cos(pitch),0,np.sin(pitch)],
        [0,1,0],
        [-np.sin(pitch),0,np.cos(pitch)]
    ])

    Rz = np.array([
        [np.cos(yaw),-np.sin(yaw),0],
        [np.sin(yaw), np.cos(yaw),0],
        [0,0,1]
    ])

    return Rz @ Ry @ Rx
```

## Transform

```python
lidar_loc = [1.5, 4.670000076293945, 4.0]
lidar_rot = [0.0, 0.0, -135.0]
```

```python
def coordinates_trans(index):
  pcd_path = os.path.join(MYDATASET_LIDAR_PATH, f"{index}.pcd")
  pcd = o3d.io.read_point_cloud(pcd_path)
  points = np.asarray(pcd.points).astype(np.float32)

  t = np.array(lidar_loc)
  R = get_rotation_matrix(lidar_rot)

  points_real = (R @ points.T).T + t

  # save new real points pcd
  pcd_real = o3d.geometry.PointCloud()
  pcd_real.points = o3d.utility.Vector3dVector(points_real)

  radar_pcd_save_path = os.path.join(MYDATASET_LIDAR_REAL_PATH, f"{index}.pcd")
  o3d.io.write_point_cloud(radar_pcd_save_path, pcd_real)
```

```python
index_range = ["017000", "017853"]
index = index_range[0]

count = 0
while index != index_range[1]:
    coordinates_trans(index)

    index = f"{int(index) + 1:06d}"
    count += 1
    if count % 10 == 0:
        print(index)
```
